```python
import matplotlib.pyplot as plt
import numpy as np
from tqdm import tqdm
```

Let $f : \mathbb{R}\to\mathbb{R}$ be a cost function defined by $$f(x)=x^2, \forall x\in\mathbb{R}.$$

```python
x = np.linspace(-1, 1)
y = x**2
plt.plot(x,y)
plt.xlabel("weight w")
plt.ylabel("cost f(w)")
plt.show()
```

**Problem** : Find $w^*\in\mathbb{R}$ such that $f(w^*) = min_{w\in\mathbb{R}} f(w)$

```python
w_init = -1.
plt.plot(w_init, w_init**2, 'ro')
plt.plot(x,y)
plt.xlabel("weight w")
plt.ylabel("cost f(w)")
plt.show()
```

**Gradient Descent Method**

$w_{new} \leftarrow w_{old} - \eta\nabla f(w_{old})$

$\eta$ is learning rate

```python
#@title Parameter setting
w_init = -1.0 #@param {type:"number"}
eta = 0.1 #@param {type:"slider", min:0, max:1, step:0.1}
epochs =  10#@param {type:"integer"}
print('initialized: {}'.format(w_init))
print('learning rate: {}'.format(eta))
print('epochs: {}'.format(epochs))
```

```python
W = np.empty(shape=(0,2))
w_old = w_init
for _ in tqdm(range(epochs)):
  W = np.concatenate((W, np.array([[w_old, w_old**2]])), axis=0)
  w_new = w_old - eta*2*w_old
  # update parameter
  w_old = w_new

plt.plot(W[:,0], W[:,1], 'r--')
plt.plot(W[:,0], W[:,1], 'ro')
plt.plot(x,y)
plt.xlabel("weight w")
plt.ylabel("cost f(w)")
plt.show()
```

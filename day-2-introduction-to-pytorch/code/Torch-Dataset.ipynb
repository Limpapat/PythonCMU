```python
import torch
from torch.utils.data import Dataset
torch.manual_seed(1)
```

## Simple Dataset

```python
class SimpleDataset(Dataset):
  def __init__(self, lenght=100):
    self.len = lenght
    self.X = 2*torch.ones(lenght, 2)
    self.y = torch.ones(lenght, 1)
  
  def __getitem__(self, index):
    sample = self.X[index], self.y[index]
    return sample
  
  def __len__(self):
    return self.len

my_dataset = SimpleDataset(lenght=10)
```

```python
X, y = my_dataset[:]
X.shape, y.shape
```

```python
X, y
```

```python
import matplotlib.pyplot as plt
plt.scatter(X.numpy(), y.numpy())
plt.show()
```

## Transforms

```python
class AddMult(object):
  def __init__(self, addx = 1, muly = 2):
    self.addx = addx
    self.muly = muly

  def __call__(self, sample):
    x, y = sample
    x += self.addx
    y *= self.muly
    return x, y

my_transform = AddMult()
```

```python
X, y = my_transform(my_dataset[:])
X.shape, y.shape
```

## Compose

```python
from torchvision import transforms
```

```python
class Mult(object):
  def __init__(self, mult = 100):
    self.mult = mult

  def __call__(self, sample):
    x, y = sample
    x *= self.mult
    y *= self.mult
    return x, y

my_transform2 = Mult()
```

```python
data_transform = transforms.Compose([my_transform, my_transform2])
```

```python
my_dataset = SimpleDataset(lenght=10)
X, y = data_transform(my_dataset[:])
X, y
```

```python
class SimpleDataset(Dataset):
  def __init__(self, lenght=100, transform=None):
    self.len = lenght
    self.X = 2*torch.ones(lenght, 2)
    self.y = torch.ones(lenght, 1)
    self.transform = transform
  
  def __getitem__(self, index):
    sample = self.X[index], self.y[index]
    if self.transform:
      sample = self.transform(sample)
    return sample
  
  def __len__(self):
    return self.len

my_dataset = SimpleDataset(lenght=10, transform=data_transform)
X, y = my_dataset[:]
X, y
```

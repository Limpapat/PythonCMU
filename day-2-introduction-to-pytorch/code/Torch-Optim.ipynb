```python
import torch
import torch.optim as optim
dir(optim)
```

```python
# Parameters Setting
w = torch.tensor([-1.], requires_grad=True) # initializing point
lr = .1                 # learning rate
epochs = 10             # epochs

optimizer = optim.SGD([w], lr=lr)
optimizer
```

```python
Ws = torch.cat((w.detach().view(-1,1), (w.detach()**2).view(-1,1)), dim=1)
Ws
```

```python
from tqdm import tqdm

for _ in tqdm(range(epochs)):
  # Compute loss
  loss = w**2
  # Reset grad & Compute grad
  optimizer.zero_grad()
  loss.backward() 
  # Update parameters
  optimizer.step()

  new_w = torch.cat((w.detach().view(-1,1), (w.detach()**2).view(-1,1)), dim=1)
  Ws = torch.cat((Ws, new_w), dim=0)
```

```python
import numpy as np
import matplotlib.pyplot as plt

dom = np.linspace(-1,1,100)
ran = dom **2

plt.plot(dom, ran)
plt.scatter(Ws.numpy()[:,0],Ws.numpy()[:,1], color='r')
plt.show()
```
